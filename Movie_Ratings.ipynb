```python
df.groupby(['Movie Names'])['Rating'].count()
```

```python
# del plt.xlabel
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(Columns1['Movie Names'], Columns1['Rating'])
plt.xlabel('Movie Name')
plt.ylabel('Rating')
plt.title('Movies_2022')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.tight_layout()
```
